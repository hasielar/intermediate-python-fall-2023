```python
import seaborn as sns
import matplotlib.pyplot as plt

class Plot:
    def __init__(self, df):
        """
        Initialize the Plot class with a DataFrame.
        
        Args:
            df (pandas.DataFrame): The input DataFrame.
        """
        self.df = df

    def box_plots(self):
        """
        Create box plots for all numeric features in the DataFrame.
        """
        numeric_features = self.df.select_dtypes(include=['number'])
        num_cols = numeric_features.columns

        num_plots = len(num_cols)
        num_rows = num_plots // 3 + 1

        plt.figure(figsize=(15, 5 * num_rows))
        for i, col in enumerate(num_cols, 1):
            plt.subplot(num_rows, 3, i)
            sns.boxplot(x=col, data=self.df)
            plt.title(f'Box Plot of {col}')
            plt.xlabel(col)

        plt.tight_layout()
        plt.show()

    def single_scatter_plot(self, x_col, y_col):
        """
        Create a single scatter plot between two columns.
        
        Args:
            x_col (str): Name of the x-axis column.
            y_col (str): Name of the y-axis column.
        """
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=x_col, y=y_col, data=self.df)
        plt.title(f'Scatter Plot between {x_col} and {y_col}')
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.show()

    def scatter_matrix_with_histograms(self):
        """
        Create a scatter matrix with histograms for numeric features.
        """
        numeric_features = self.df.select_dtypes(include=['number'])
        num_cols = numeric_features.columns

        plt.figure(figsize=(12, 12))
        sns.set(style='ticks')
        sns.pairplot(self.df, vars=num_cols, diag_kind='kde')

        plt.suptitle('Scatter Matrix with Histograms', y=1.02)
        plt.show()

# Demonstrate the usage on the Titanic dataset
import pandas as pd

# Load the Titanic dataset (you can replace this with your own dataset)
titanic_df = pd.read_csv('data/titanic.csv')

# Create an instance of the Plot class
plotter = Plot(titanic_df)

# Create box plots for numeric features
plotter.box_plots()

# Create a single scatter plot between 'Age' and 'Fare'
plotter.single_scatter_plot('Age', 'Fare')

# Create a scatter matrix with histograms
plotter.scatter_matrix_with_histograms()
```
